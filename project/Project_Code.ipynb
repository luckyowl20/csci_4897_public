```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# --- 1. Model Setup and Parameters ---

#  Population Structure
N_ages = 3  
age_labels = ['Children (0-17)', 'Adults (18-64)', 'Elderly (65+)']
compartment_labels = ['Susceptible (S)', 'Exposed (E)', 'Infectious (I)', 'Recovered (R)', 'Deceased (D)']
N_initial = np.array([1246100, 3757100, 953500]) # Absolute Initial Population Sizes
Total_Population_initial = np.sum(N_initial)

# Epidemiological Parameters
beta = 0.05      
sigma = 0.1   
gamma = 0.025  
mu_age = np.array([0.005, 0.002, 0.02]) 

# Contact Matrix (C) for no social distancing
C = np.array([
    [10.0, 5.0, 2.0], 
    [5.0, 15.0, 3.0], 
    [2.0, 3.0, 6.0]   
])

t_span = (0, 300)
t_eval = np.linspace(t_span[0], t_span[1], 301)


# --- 2. THE CORE SEIRD Model ODE System (Corrected Indexing) ---

def seird_model(t, y, N_ages, N_initial, C, beta, sigma, gamma, mu_age,
                vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                t_anti_start, t_dist_start):
    """
    Age-structured SEIRD model with correct handling of age-group compartments.
    y is a flat vector of length N_ages * 5.
    """
    
    # Compartments S, E, I, R, D are indexed 0, 1, 2, 3, 4 respectively.
    # We slice the flat vector y to get the current state for each compartment across all age groups.
    S = y[0*N_ages : 1*N_ages]
    E = y[1*N_ages : 2*N_ages]
    I = y[2*N_ages : 3*N_ages]
    R = y[3*N_ages : 4*N_ages]
    D = y[4*N_ages : 5*N_ages]

    # Calculate the current living population for each group (S+E+I+R)
    N_current = S + E + I + R

    # 1. Interventions (now time-dependent)

    # Vaccination: still constant in time (you can time-gate later if you want)
    vax_flow_rate = vax_rate * vax_efficacy   # fraction per day * efficacy

    # --- Antivirals: turn on at t_anti_start ---
    if t < t_anti_start:
        antiviral_eff_t = np.zeros_like(antiviral_efficacy)   # no antivirals yet
    else:
        antiviral_eff_t = antiviral_efficacy                  # antivirals active

    # Antivirals reduce mortality: effective mortality = mu_age * (1 - efficacy)
    mu_adj = mu_age * (1.0 - antiviral_eff_t)

    # --- Social distancing: turn on at t_dist_start ---
    if t < t_dist_start:
        dist_factor_vector = np.ones_like(distancing_factor)  # no distancing yet
    else:
        # distancing_factor is "fraction of contacts removed" for each age group
        dist_factor_vector = 1.0 - distancing_factor

    # Modify contact matrix with age-specific distancing
    dist_matrix = np.outer(dist_factor_vector, dist_factor_vector)
    C_adj = C * dist_matrix

    # Force of Infection (Lambda_i)
    Lambda = np.zeros(N_ages)
    for i in range(N_ages): # For susceptible group i
        exposure_from_j = 0.0
        for j in range(N_ages): # From infectious group j
            # I[j] / N_safe[j] ensures correct pairing of infected count and its population base.
            exposure_from_j += C_adj[i, j] * (I[j] / N_current[j])
        
        # Susceptibility factor is 1.0 (as vaccinated moved to R at t=0)
        Lambda[i] = beta * exposure_from_j
        
    # ODE System (returns dy/dt as a flat vector)
    dydt = np.zeros(N_ages * 5)
    
    for i in range(N_ages):
        dSi_dt = -Lambda[i] * S[i] -vax_flow_rate[i] * S[i] 
        dEi_dt = Lambda[i] * S[i] - sigma * E[i]
        dIi_dt = sigma * E[i] - gamma * I[i] - mu_adj[i] * I[i]
        dRi_dt = gamma * I[i] + vax_flow_rate[i] * S[i] 
        dDi_dt = mu_adj[i] * I[i]
        
        # Populate the flat dydt vector correctly
        dydt[0*N_ages + i] = dSi_dt
        dydt[1*N_ages + i] = dEi_dt
        dydt[2*N_ages + i] = dIi_dt
        dydt[3*N_ages + i] = dRi_dt
        dydt[4*N_ages + i] = dDi_dt
        
    return dydt


# --- 3. Simulation Function and Scenario Definition ---

def run_simulation(vax_rate_target, vax_efficacy,
                   antiviral_efficacy, distancing_factor,
                   label,
                   t_anti_start=0.0,   # default: antivirals on from t=0
                   t_dist_start=0.0):  # default: distancing on from t=0
    """Runs one SEIRD simulation with initial condition adjustment for vaccination."""

    # 3a. Base Initial Conditions
    I0 = np.zeros(N_ages); I0[1] = 1.0 # 1 infected Adult
    E0 = np.zeros(N_ages); R0 = np.zeros(N_ages); D0 = np.zeros(N_ages)
    S0 = N_initial - I0 - E0 - R0 - D0 
    
    # Flatten the Initial Conditions (IC) vector: S S S | E E E | I I I | R R R | D D D
    IC_adjusted = np.concatenate([S0, E0, I0, R0, D0])
    
    # 3c. Solve ODE
    solver_args = (N_ages, N_initial, C, beta, sigma, gamma, mu_age,
                   vax_rate_target, vax_efficacy,
                   antiviral_efficacy, distancing_factor,
                   t_anti_start, t_dist_start)
    
    sol = solve_ivp(
        fun=seird_model, t_span=t_span, y0=IC_adjusted, t_eval=t_eval,
        args=solver_args, method='RK45'
    )
    
    # The output is (N_time x N_groups*5). We now reshape it to (N_time x 5 compartments x N_groups)
    # The previous method using (N_time x N_groups x 5 compartments) was prone to the bug.
    results_abs_flat = sol.y.T
    results_abs_reshaped = np.zeros((len(t_eval), N_ages, 5))

    # CRITICAL FIX: Manually assign the results to the correct [time, group, compartment] indices
    # This prevents the mixing of compartments that caused the I/N > 1 error.
    for k in range(5): # k = compartment index (0=S, 4=D)
        results_abs_reshaped[:, :, k] = results_abs_flat[:, k*N_ages : (k+1)*N_ages]
        
    results_prop_total = results_abs_reshaped.sum(axis=1)[:, 4] / Total_Population_initial # Only Deaths/N_total shown for conciseness

    return {
        't': sol.t, 
        'abs': results_abs_reshaped, # Shape: (Time x Age Group x Compartment)
        'prop_total': results_prop_total,
        'label': label
    }

# --- Define Scenarios ---

# Scenario A: Baseline (No Intervention)
vax_A  = np.zeros(N_ages)
anti_A = np.zeros(N_ages)
dist_A = np.zeros(N_ages)

results_A = run_simulation(
    vax_A, 0.0, anti_A, dist_A,
    label="Baseline",
    t_anti_start=300.0,  # put start times far in the future to effectively "never"
    t_dist_start=300.0
)

# --- Scenario B-REV: Targeted Combined Intervention ---

# Vaccination: moderate daily vaccination across ages
vax_B_rev    = np.array([0.01, 0.005, 0.015])  # children, adults, elderly
vax_eff_B_rev = 0.8                            # 80% vaccine efficacy

# Antivirals: mainly for elderly (reduce death rate)
anti_B_rev   = np.array([0.0, 0.0, 0.5])       # 50% antiviral effect in elderly

# Social distancing: mainly among adults (reduce contacts)
dist_B_rev   = np.array([0.0, 0.2, 0.0])       # 20% distancing in adults only

# Timing of interventions
t_dist_start_B = 30.0   # distancing starts at day 30
t_anti_start_B = 60.0   # antivirals start at day 60

results_B_rev = run_simulation(
    vax_B_rev,           # vax_rate_target
    vax_eff_B_rev,       # vax_efficacy
    anti_B_rev,          # antiviral_efficacy (per age group)
    dist_B_rev,          # distancing_factor (per age group)
    "Targeted Combined Intervention",
    t_anti_start=t_anti_start_B,
    t_dist_start=t_dist_start_B
)



# --- 4. Visualization Functions (Generates all 5 requested figures) ---

Compartment_Names = ['Susceptible (S)', 'Exposed (E)', 'Infectious (I)', 'Recovered (R)', 'Deaths (D)']
Colors_comp = ['#2ca02c', '#ff7f0e', '#d62728', '#1f77b4', '#7f7f7f']
Colors_age = ['#2ca02c', '#ff7f0e', '#d62728']

# --- Figure 1: Proportions of TOTAL Population ---
def plot_proportions_total(results_A, results_B):
    fig, (ax1_A, ax1_B) = plt.subplots(1, 2, figsize=(16, 6))
    
    # We must calculate the proportion of total N for all compartments here
    for ax, results_set in zip([ax1_A, ax1_B], [results_A, results_B]):
        total_comp_abs = results_set['abs'].sum(axis=1) # Sums across age groups -> (Time x 5 Compartments)
        total_comp_prop = total_comp_abs / Total_Population_initial
        
        ax.set_title(f'{results_set["label"]} - Total Population Dynamics', fontsize=12)
        ax.set_xlabel('Time (Days)'); ax.set_ylabel('Proportion of Total Initial Population')
        for k in range(5):
            ax.plot(results_set['t'], total_comp_prop[:, k], label=Compartment_Names[k], color=Colors_comp[k])
        ax.legend(loc='center right', fontsize=8); ax.grid(True, linestyle=':', alpha=0.6)
    
    fig.suptitle('Figure 1: Comparison of Overall Epidemic Dynamics (Proportions of Total N)', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

# --- Figures 2 & 3: Absolute Numbers (I and D) ---
def plot_absolute_details(results_set, fig_num):
    fig, (ax_main, ax_death) = plt.subplots(2, 1, figsize=(8, 10))
    # Top Subplot: Infectious (I) Absolute Numbers
    ax_main.set_title(f'{results_set["label"]} - Absolute Infectious (I) by Age Group', fontsize=12)
    ax_main.set_ylabel('Number of Individuals (Infectious)'); ax_main.set_xlabel('Time (Days)')
    for i in range(N_ages):
        ax_main.plot(results_set['t'], results_set['abs'][:, i, 2], label=age_labels[i], linestyle='-', linewidth=2, color=Colors_age[i])
    ax_main.legend(loc='upper right', fontsize=8); ax_main.grid(True, linestyle=':', alpha=0.6)
    # Bottom Subplot: Cumulative Deaths (D) Absolute Numbers
    ax_death.set_title('Cumulative Deaths (D) by Age Group', fontsize=10)
    ax_death.set_ylabel('Number of Deaths'); ax_death.set_xlabel('Time (Days)')
    for i in range(N_ages):
        ax_death.plot(results_set['t'], results_set['abs'][:, i, 4], label=age_labels[i], linestyle='--', linewidth=1.5, color=Colors_age[i])
    ax_death.legend(loc='lower right', fontsize=8); ax_death.grid(True, linestyle=':', alpha=0.6)
    fig.suptitle(f'Figure {fig_num}: {results_set["label"]} - Absolute Burden by Age', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

# --- Figures 4 & 5: Age-Specific Proportions (I/N_i and D/N_i) ---
def plot_age_proportions(results_set, N_pop_initial, fig_num):
    """Plots the proportion of each age group that is Infectious (I) and Dead (D)."""
    # The shape of results_set['abs'] is (Time x Age Group x Compartment)
    # The N_pop_initial must be broadcast correctly across the time axis.
    I_prop = results_set['abs'][:, :, 2] / N_pop_initial 
    D_prop = results_set['abs'][:, :, 4] / N_pop_initial

    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Subplot 1: Infectious (I) Proportion
    ax1 = axes[0]
    ax1.set_title(f'Peak Infection by Age Group (Proportion of $N_i$)', fontsize=14)
    ax1.set_xlabel('Time (Days)'); ax1.set_ylabel('Proportion of Age Group Infected ($I / N_i$)')
    for i in range(N_ages):
        ax1.plot(results_set['t'], I_prop[:, i], label=age_labels[i], linestyle='-', linewidth=2, color=Colors_age[i])
    ax1.legend(loc='upper right', fontsize=10); ax1.set_ylim(0, 1.05); ax1.grid(True, linestyle=':', alpha=0.6)
    
    # Subplot 2: Cumulative Deaths (D) Proportion
    ax2 = axes[1]
    ax2.set_title(f'Final Mortality by Age Group (Proportion of $N_i$)', fontsize=14)
    ax2.set_xlabel('Time (Days)'); ax2.set_ylabel('Proportion of Age Group Dead ($D / N_i$)')
    for i in range(N_ages):
        ax2.plot(results_set['t'], D_prop[:, i], label=age_labels[i], linestyle='--', linewidth=2, color=Colors_age[i])
    ax2.legend(loc='lower right', fontsize=10); ax2.grid(True, linestyle=':', alpha=0.6)
    
    fig.suptitle(f'Figure {fig_num}: {results_set["label"]} - Age-Specific Disease Burden (Proportions)', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()


# --- 5. Generate All Requested Graphs ---

# Figure 1: Overall Dynamics (Proportion of Total Population)
plot_proportions_total(results_A, results_B_rev)

# Figure 2: Baseline Absolute Numbers
plot_absolute_details(results_A, 2)

# Figure 3: Targeted Absolute Numbers
plot_absolute_details(results_B_rev, 3)

# Figure 4: Baseline Age-Specific Proportions (I/N_i and D/N_i)
plot_age_proportions(results_A, N_initial, 4)

# Figure 5: Targeted Age-Specific Proportions (I/N_i and D/N_i)
plot_age_proportions(results_B_rev, N_initial, 5)

print("\n--- Summary of Scenarios ---")
print("Scenario A (Baseline): No interventions.")
print("Scenario B (Targeted): Vax (Adults: 20% coverage * 80% eff = 16% moved from S to R at t=0), Antivirals (Elderly, 50% eff), Distancing (Adults, 20% reduction).")
```

### Sensitivity analysis for vaccines

```python

def seird_model(t, y, N_ages, N_initial, C, beta, sigma, gamma, mu_age,
                vax_rate_flow, vax_efficacy, antiviral_efficacy, distancing_factor, vax_start_time):
    """
    Age-structured SEIRD model with dynamic population base and time-dependent vaccination.
    vax_rate_flow is now the constant per-capita rate applied after vax_start_time.
    """
    
    S = y[0*N_ages : 1*N_ages]
    E = y[1*N_ages : 2*N_ages]
    I = y[2*N_ages : 3*N_ages]
    R = y[3*N_ages : 4*N_ages]
    D = y[4*N_ages : 5*N_ages]

    N_current = S + E + I + R

    # --- Time-Dependent Vaccination Rate ---
    vax_flow_rate = np.zeros(N_ages)
    if t >= vax_start_time:
        # Rate of flow from S to R (per person per day)
        # Note: We assume the flow rate accounts for efficacy: flow = vax_rate * vax_efficacy
        vax_flow_rate = vax_rate_flow * vax_efficacy 

    # 1. Interventions
    mu_adj = mu_age * (1.0 - antiviral_efficacy)
    dist_factor_vector = 1.0 - distancing_factor
    dist_matrix = np.outer(dist_factor_vector, dist_factor_vector)
    C_adj = C * dist_matrix 

    # Force of Infection (Lambda_i)
    Lambda = np.zeros(N_ages)
    for i in range(N_ages): 
        exposure_from_j = 0.0
        for j in range(N_ages): 
            exposure_from_j += C_adj[i, j] * (I[j] / N_current[j])
        Lambda[i] = beta * exposure_from_j
        
    # ODE System 
    dydt = np.zeros(N_ages * 5)
    
    for i in range(N_ages):
        dSi_dt = -Lambda[i] * S[i] - vax_flow_rate[i] * S[i]
        dEi_dt = Lambda[i] * S[i] - sigma * E[i]
        dIi_dt = sigma * E[i] - gamma * I[i] - mu_adj[i] * I[i]
        dRi_dt = gamma * I[i] + vax_flow_rate[i] * S[i]  # Vaccination flow adds to R
        dDi_dt = mu_adj[i] * I[i]
        
        dydt[0*N_ages + i] = dSi_dt
        dydt[1*N_ages + i] = dEi_dt
        dydt[2*N_ages + i] = dIi_dt
        dydt[3*N_ages + i] = dRi_dt
        dydt[4*N_ages + i] = dDi_dt
        
    return dydt


# --- 3. Simulation Function (Simplified IC) ---

def run_simulation(vax_rate_flow, vax_efficacy, antiviral_efficacy, distancing_factor, vax_start_time):
    """Runs one SEIRD simulation and returns time, absolute results, and label."""
    
    # Simple Initial Conditions (IC) setup: No t=0 vax
    I0 = np.zeros(N_ages); I0[1] = 1.0 
    E0 = np.zeros(N_ages); R0 = np.zeros(N_ages); D0 = np.zeros(N_ages)
    S0 = N_initial - I0 - E0 - R0 - D0 
    IC = np.concatenate([S0, E0, I0, R0, D0])
    
    solver_args = (N_ages, N_initial, C, beta, sigma, gamma, mu_age,
                   vax_rate_flow, vax_efficacy, antiviral_efficacy, distancing_factor, vax_start_time)
    
    sol = solve_ivp(
        fun=seird_model, t_span=t_span, y0=IC, t_eval=t_eval,
        args=solver_args, method='RK45'
    )
    
    results_abs_flat = sol.y.T
    results_abs_reshaped = np.zeros((len(t_eval), N_ages, 5))
    for k in range(5):
        results_abs_reshaped[:, :, k] = results_abs_flat[:, k*N_ages : (k+1)*N_ages]

    return results_abs_reshaped


# --- 4. Sensitivity Analysis: Varying Vaccination Start Time ---

def sensitivity_analysis_vax_time(vax_rate, vax_eff, anti_eff, dist_factor):
    """
    Runs simulations for a range of vaccination start times (t_vax)
    and collects peak infections and final deaths.
    """
    # Range of Vax Start Times (Days 0 to 200)
    vax_start_times = np.arange(0, 201, 25) 
    
    # Results storage: Peak I and Final D for each group, for each start time.
    # Shape: (N_start_times x N_ages)
    peak_I_by_group = np.zeros((len(vax_start_times), N_ages))
    final_D_by_group = np.zeros((len(vax_start_times), N_ages))

    print("Running Sensitivity Analysis (Varying Vax Start Time)...")

    for idx, t_vax in enumerate(vax_start_times):
        # Run simulation with fixed intervention parameters but varying start time
        results = run_simulation(vax_rate, vax_eff, anti_eff, dist_factor, t_vax)
        
        # Calculate Peak I (max across time, for each group)
        peak_I_by_group[idx, :] = np.max(results[:, :, 2], axis=0)
        
        # Calculate Final D (last value in time, for each group)
        final_D_by_group[idx, :] = results[-1, :, 4]
        
    return vax_start_times, peak_I_by_group, final_D_by_group

# Define fixed intervention scenario (Adults are targeted)
# Vax rate is the per-capita rate: 0.025/day (achieving ~50% coverage over 20 days if S stays constant)
vax_rate_flow_target = np.array([0.01, 0.005, 0.01])  
vax_eff_target = 0.8                   
anti_eff_target = np.array([0.0, 0.0, 0.5]) 
dist_factor_target = np.array([0.0, 0.2, 0.0]) 

vax_rate_flow_target_none = np.array([0,0,0])  

# Run the analysis
t_vax_range, peak_I, final_D = sensitivity_analysis_vax_time(
    vax_rate_flow_target, vax_eff_target, anti_eff_target, dist_factor_target
)

t_vax_range_none, peak_I_none, final_D_none = sensitivity_analysis_vax_time(
    vax_rate_flow_target_none, vax_eff_target, anti_eff_target, dist_factor_target
)


# --- 5. Visualization of Sensitivity Analysis ---

Colors_age = ['#2ca02c', '#ff7f0e', '#d62728']

def plot_sensitivity_results(t_range, peak_I, final_D):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # --- Subplot 1: Peak Infectious (I) ---
    ax1 = axes[0]
    ax1.set_title('Impact of Vaccination Timing on Peak Infections ($I$)', fontsize=14)
    ax1.set_xlabel('Vaccination Introduction Time ($t_{vax}$, Days)')
    ax1.set_ylabel('Peak Absolute Infections (max $I_i$)')
    
    for i in range(N_ages):
        ax1.plot(t_range, peak_I[:, i], 
                 label=age_labels[i], marker='o', linestyle='-', color=Colors_age[i])
        
    ax1.legend(loc='upper right', fontsize=10)
    ax1.grid(True, linestyle=':', alpha=0.6)
    
    # --- Subplot 2: Total Deaths (D) ---
    ax2 = axes[1]
    ax2.set_title('Impact of Vaccination Timing on Total Deaths ($D$)', fontsize=14)
    ax2.set_xlabel('Vaccination Introduction Time ($t_{vax}$, Days)')
    ax2.set_ylabel('Total Cumulative Deaths ($D$)')
    
    for i in range(N_ages):
        ax2.plot(t_range, final_D[:, i], 
                 label=age_labels[i], marker='s', linestyle='--', color=Colors_age[i])
        
    ax2.legend(loc='upper right', fontsize=10)
    ax2.grid(True, linestyle=':', alpha=0.6)
    
    fig.suptitle('Sensitivity Analysis: Varying Vaccination Time of Introduction', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Run the visualization
plot_sensitivity_results(t_vax_range, peak_I, final_D)

print("\n--- Summary of Analysis ---")
print(f"Analysis shows the impact of delaying vaccination (Adults targeted at rate 0.005/day, Elderly and Children target at rate of 0.01/day, 80% efficacy) on peak infections and total deaths.")
print("The earlier the vaccination starts, the lower the peak infections and total deaths are across all age groups.")

plot_sensitivity_results(t_vax_range_none, peak_I_none, final_D_none)
```

### Sensitivity analysis for all 3 interventions without cost

```python
# ============================================================
# NEW CELL: Joint Timing Optimization for Vax, Antivirals, Distancing
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp

# 1) SEIRD model with three time-dependent interventions
def seird_model_timing_opt(t, y,
                           N_ages, N_initial, C, beta, sigma, gamma, mu_age,
                           vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                           t_vax_start, t_anti_start, t_dist_start):
    """
    Age-structured SEIRD model with time-dependent:
      - vaccination start time (t_vax_start),
      - antiviral start time (t_anti_start),
      - distancing start time (t_dist_start).

    vax_rate, antiviral_efficacy, distancing_factor are per-age arrays.
    """

    # Unpack compartments
    S = y[0*N_ages : 1*N_ages]
    E = y[1*N_ages : 2*N_ages]
    I = y[2*N_ages : 3*N_ages]
    R = y[3*N_ages : 4*N_ages]
    D = y[4*N_ages : 5*N_ages]

    N_current = S + E + I + R
    N_safe = np.maximum(N_current, 1e-12)  # avoid division by zero

    # --- Vaccination: activates at t_vax_start ---
    if t >= t_vax_start:
        # per-capita rate * efficacy
        vax_flow_rate = vax_rate * vax_efficacy
    else:
        vax_flow_rate = np.zeros_like(vax_rate)

    # --- Antivirals: reduce mortality after t_anti_start ---
    if t >= t_anti_start:
        antiviral_eff_t = antiviral_efficacy
    else:
        antiviral_eff_t = np.zeros_like(antiviral_efficacy)
    mu_adj = mu_age * (1.0 - antiviral_eff_t)

    # --- Distancing: reduces contacts after t_dist_start ---
    if t >= t_dist_start:
        dist_factor_vector = 1.0 - distancing_factor  # fraction of contacts remaining
    else:
        dist_factor_vector = np.ones_like(distancing_factor)

    dist_matrix = np.outer(dist_factor_vector, dist_factor_vector)
    C_adj = C * dist_matrix

    # --- Force of infection ---
    Lambda = np.zeros(N_ages)
    for i in range(N_ages):  # susceptible group i
        exposure_from_j = 0.0
        for j in range(N_ages):  # infectious group j
            exposure_from_j += C_adj[i, j] * (I[j] / N_safe[j])
        Lambda[i] = beta * exposure_from_j

    # --- ODE system ---
    dydt = np.zeros(N_ages * 5)
    for i in range(N_ages):
        dSi_dt = -Lambda[i] * S[i] - vax_flow_rate[i] * S[i]
        dEi_dt =  Lambda[i] * S[i] - sigma * E[i]
        dIi_dt =  sigma * E[i] - gamma * I[i] - mu_adj[i] * I[i]
        dRi_dt =  gamma * I[i] + vax_flow_rate[i] * S[i]
        dDi_dt =  mu_adj[i] * I[i]

        dydt[0*N_ages + i] = dSi_dt
        dydt[1*N_ages + i] = dEi_dt
        dydt[2*N_ages + i] = dIi_dt
        dydt[3*N_ages + i] = dRi_dt
        dydt[4*N_ages + i] = dDi_dt

    return dydt


# 2) Run a single simulation for a given triple of start times
def run_single_timing_sim(t_vax_start, t_anti_start, t_dist_start,
                          vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor):
    """
    Runs one SEIRD simulation with the given intervention start times.
    Returns (t, results_abs_reshaped) where results_abs_reshaped has shape (Time x Age x Compartment).
    Assumes global N_ages, N_initial, C, beta, sigma, gamma, mu_age, t_span, t_eval are defined.
    """
    # Initial conditions: 1 infected adult, rest susceptible
    I0 = np.zeros(N_ages); I0[1] = 1.0
    E0 = np.zeros(N_ages); R0 = np.zeros(N_ages); D0 = np.zeros(N_ages)
    S0 = N_initial - I0 - E0 - R0 - D0
    y0 = np.concatenate([S0, E0, I0, R0, D0])

    solver_args = (
        N_ages, N_initial, C, beta, sigma, gamma, mu_age,
        vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
        t_vax_start, t_anti_start, t_dist_start
    )

    sol = solve_ivp(
        fun=seird_model_timing_opt,
        t_span=t_span,
        y0=y0,
        t_eval=t_eval,
        args=solver_args,
        method='RK45'
    )

    results_abs_flat = sol.y.T  # (Time x [Age*5])
    results_abs_reshaped = np.zeros((len(t_eval), N_ages, 5))
    for k in range(5):
        results_abs_reshaped[:, :, k] = results_abs_flat[:, k*N_ages : (k+1)*N_ages]

    return sol.t, results_abs_reshaped


# 3) Evaluate objective for one combination (minimize final total deaths)
def evaluate_timing_combo(t_vax_start, t_anti_start, t_dist_start,
                          vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor):
    """
    Runs a single simulation and returns:
      - final_total_deaths (objective),
      - peak_total_infections (useful secondary metric).
    """
    t, results = run_single_timing_sim(
        t_vax_start, t_anti_start, t_dist_start,
        vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor
    )

    # results: (Time x Age x Compartment)
    total_I = results[:, :, 2].sum(axis=1)  # Infectious across all ages
    total_D = results[:, :, 4].sum(axis=1)  # Deaths across all ages

    final_total_deaths = total_D[-1]
    peak_total_infections = total_I.max()

    return final_total_deaths, peak_total_infections


# 4) Grid search over all three start times
def joint_timing_grid_search(vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                             t_vax_values, t_anti_values, t_dist_values):
    """
    Explores a 3D grid of start times:
      t_vax_values   : array of vaccination start times (days)
      t_anti_values  : array of antiviral start times (days)
      t_dist_values  : array of distancing start times (days)

    Returns:
      best_result: dict with best combo and metrics
      deaths_grid: 3D array of final total deaths for each combo
    """
    deaths_grid = np.zeros((len(t_vax_values), len(t_anti_values), len(t_dist_values)))

    best = {
        "t_vax_start": None,
        "t_anti_start": None,
        "t_dist_start": None,
        "final_deaths": np.inf,
        "peak_infections": np.inf
    }

    for i, tv in enumerate(t_vax_values):
        for j, ta in enumerate(t_anti_values):
            for k, td in enumerate(t_dist_values):
                final_deaths, peak_I = evaluate_timing_combo(
                    tv, ta, td,
                    vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor
                )

                deaths_grid[i, j, k] = final_deaths

                # Main objective: minimize final deaths
                if final_deaths < best["final_deaths"]:
                    best.update({
                        "t_vax_start": tv,
                        "t_anti_start": ta,
                        "t_dist_start": td,
                        "final_deaths": final_deaths,
                        "peak_infections": peak_I
                    })

    return best, deaths_grid


# 5) Example usage: define intervention intensities and timing grid, then run the search

# Use your "Scenario B" style intensities as the fixed strengths:
vax_rate_opt    = np.array([0.01, 0.005, 0.015])  # per-day vax rate by age
vax_eff_opt     = 0.8                             # vaccine efficacy (scalar)
anti_eff_opt    = np.array([0.0, 0.0, 0.5])       # antivirals mainly for elderly
dist_factor_opt = np.array([0.0, 0.2, 0.0])       # distancing mainly for adults

# Grids of start times (you can refine these later)
t_vax_values   = np.arange(0, 181, 30)   # 0, 30, 60, 90, 120, 150, 180
t_anti_values  = np.arange(0, 181, 30)
t_dist_values  = np.arange(0, 181, 30)

best_combo, deaths_grid = joint_timing_grid_search(
    vax_rate_opt, vax_eff_opt, anti_eff_opt, dist_factor_opt,
    t_vax_values, t_anti_values, t_dist_values
)

print("\n=== Best Timing Combination (min final total deaths) ===")
print(f"Vaccination start time: {best_combo['t_vax_start']} days")
print(f"Antiviral start time : {best_combo['t_anti_start']} days")
print(f"Distancing start time: {best_combo['t_dist_start']} days")
print(f"Final total deaths   : {best_combo['final_deaths']:.0f}")
print(f"Peak total infections: {best_combo['peak_infections']:.0f}")

# Optionally, re-run the best scenario to inspect full trajectories:
t_best, results_best = run_single_timing_sim(
    best_combo['t_vax_start'],
    best_combo['t_anti_start'],
    best_combo['t_dist_start'],
    vax_rate_opt, vax_eff_opt, anti_eff_opt, dist_factor_opt
)
# 't_best' and 'results_best' can now be passed to your plotting functions if you like.

# 6) Visualization: Show epidemic curves with best start times + bar chart

def plot_best_timing_summary(t, results, best_combo):
    """
    t: array of time points
    results: (Time x Age x Compartment)
    best_combo: dict from joint_timing_grid_search with optimal start times
    """
    total_I = results[:, :, 2].sum(axis=1)  # total infectious over time
    total_D = results[:, :, 4].sum(axis=1)  # total deaths over time

    fig, axes = plt.subplots(1, 1, figsize=(16, 6))

    # --- Left subplot: epidemic curves + vertical lines at best times ---
    ax1 = axes
    ax1.plot(t, total_I, label='Total Infectious (I)', linewidth=2)
    ax1.plot(t, total_D, label='Total Deaths (D)', linewidth=2)

    # Draw vertical lines for best start times
    ax1.axvline(best_combo['t_vax_start'],  linestyle='--', alpha=0.8, label=f"Vax start = {best_combo['t_vax_start']} d")
    ax1.axvline(best_combo['t_anti_start'], linestyle='--', alpha=0.8, label=f"Antivirals start = {best_combo['t_anti_start']} d")
    ax1.axvline(best_combo['t_dist_start'], linestyle='--', alpha=0.8, label=f"Distancing start = {best_combo['t_dist_start']} d")

    ax1.set_title('Best-Timing Scenario: Epidemic Curves & Intervention Onset', fontsize=12)
    ax1.set_xlabel('Time (days)')
    ax1.set_ylabel('Number of individuals')
    ax1.legend(loc='upper left', fontsize=8)
    ax1.grid(True, linestyle=':', alpha=0.6)

    fig.suptitle('Optimal Joint Timing of Vaccination, Antivirals, and Distancing', fontsize=14)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    plt.savefig("optimal_joint_timing.png")


# Call the plotting function for the best combo
plot_best_timing_summary(t_best, results_best, best_combo)

```

### adding in cost

```python
# ============================================================
# NEW CELL: Joint Timing Optimization with Costs
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 1) SEIRD model with three time-dependent interventions + cost state
def seird_model_timing_opt(t, y,
                           N_ages, N_initial, C, beta, sigma, gamma, mu_age,
                           vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                           t_vax_start, t_anti_start, t_dist_start,
                           cost_vax, cost_anti, cost_dist):
    """
    Age-structured SEIRD model with time-dependent:
      - vaccination start time (t_vax_start),
      - antiviral start time (t_anti_start),
      - distancing start time (t_dist_start),
    and an additional state C(t) for cumulative cost.

    vax_rate, antiviral_efficacy, distancing_factor are per-age arrays.
    cost_vax, cost_anti, cost_dist are per-age cost parameters:
      - cost_vax[i]: $ per vaccinated person in age group i
      - cost_anti[i]: $ per treated infectious person in age group i
      - cost_dist[i]: $ per person-day at full distancing in age group i
    """

    # Unpack compartments
    S = y[0*N_ages : 1*N_ages]
    E = y[1*N_ages : 2*N_ages]
    I = y[2*N_ages : 3*N_ages]
    R = y[3*N_ages : 4*N_ages]
    D = y[4*N_ages : 5*N_ages]
    C_cost = y[5*N_ages]  # cumulative cost

    N_current = S + E + I + R
    N_safe = np.maximum(N_current, 1e-12)  # avoid division by zero

    # --- Vaccination: activates at t_vax_start ---
    if t >= t_vax_start:
        # per-capita rate * efficacy
        vax_flow_rate = vax_rate * vax_efficacy
    else:
        vax_flow_rate = np.zeros_like(vax_rate)

    # --- Antivirals: reduce mortality after t_anti_start ---
    if t >= t_anti_start:
        antiviral_eff_t = antiviral_efficacy
    else:
        antiviral_eff_t = np.zeros_like(antiviral_efficacy)
    mu_adj = mu_age * (1.0 - antiviral_eff_t)

    # --- Distancing: reduces contacts after t_dist_start ---
    if t >= t_dist_start:
        dist_factor_vector = 1.0 - distancing_factor  # fraction of contacts remaining
        active_distancing = True
    else:
        dist_factor_vector = np.ones_like(distancing_factor)
        active_distancing = False

    dist_matrix = np.outer(dist_factor_vector, dist_factor_vector)
    C_adj = C * dist_matrix

    # --- Force of infection ---
    Lambda = np.zeros(N_ages)
    for i in range(N_ages):  # susceptible group i
        exposure_from_j = 0.0
        for j in range(N_ages):  # infectious group j
            exposure_from_j += C_adj[i, j] * (I[j] / N_safe[j])
        Lambda[i] = beta * exposure_from_j

    # --- ODE system for epidemiology ---
    dydt = np.zeros(N_ages * 5 + 1)
    for i in range(N_ages):
        dSi_dt = -Lambda[i] * S[i] - vax_flow_rate[i] * S[i]
        dEi_dt =  Lambda[i] * S[i] - sigma * E[i]
        dIi_dt =  sigma * E[i] - gamma * I[i] - mu_adj[i] * I[i]
        dRi_dt =  gamma * I[i] + vax_flow_rate[i] * S[i]
        dDi_dt =  mu_adj[i] * I[i]

        dydt[0*N_ages + i] = dSi_dt
        dydt[1*N_ages + i] = dEi_dt
        dydt[2*N_ages + i] = dIi_dt
        dydt[3*N_ages + i] = dRi_dt
        dydt[4*N_ages + i] = dDi_dt

    # --- Cost dynamics: dC/dt ---
    # Vaccination cost: cost per vaccinated person * flow of vaccinated individuals
    vax_flow = vax_flow_rate * S  # persons/day
    cost_rate_vax = np.sum(cost_vax * vax_flow)

    # Antiviral cost: cost per treated infectious person * treated flow
    # Approximate treated per day as antiviral_eff_t * I (fraction benefiting from antivirals)
    anti_treated = antiviral_eff_t * I
    cost_rate_anti = np.sum(cost_anti * anti_treated)

    # Distancing cost: cost per person-day (at full distancing) * distancing intensity * population
    if active_distancing:
        cost_rate_dist = np.sum(cost_dist * distancing_factor * N_current)
    else:
        cost_rate_dist = 0.0

    dC_dt = cost_rate_vax + cost_rate_anti + cost_rate_dist
    dydt[5*N_ages] = dC_dt

    return dydt


# 2) Run a single simulation for a given triple of start times
def run_single_timing_sim(t_vax_start, t_anti_start, t_dist_start,
                          vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                          cost_vax, cost_anti, cost_dist):
    """
    Runs one SEIRD simulation with the given intervention start times.
    Returns (t, results_abs_reshaped, total_cost) where:
      - results_abs_reshaped has shape (Time x Age x Compartment),
      - total_cost is the cumulative cost C(T).
    Assumes global N_ages, N_initial, C, beta, sigma, gamma, mu_age, t_span, t_eval are defined.
    """
    # Initial conditions: 1 infected adult, rest susceptible, zero cost
    I0 = np.zeros(N_ages); I0[1] = 1.0
    E0 = np.zeros(N_ages); R0 = np.zeros(N_ages); D0 = np.zeros(N_ages)
    S0 = N_initial - I0 - E0 - R0 - D0
    C0 = np.array([0.0])  # initial cost
    y0 = np.concatenate([S0, E0, I0, R0, D0, C0])

    solver_args = (
        N_ages, N_initial, C, beta, sigma, gamma, mu_age,
        vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
        t_vax_start, t_anti_start, t_dist_start,
        cost_vax, cost_anti, cost_dist
    )

    sol = solve_ivp(
        fun=seird_model_timing_opt,
        t_span=t_span,
        y0=y0,
        t_eval=t_eval,
        args=solver_args,
        method='RK45'
    )

    results_abs_flat = sol.y.T  # (Time x [Age*5 + 1])
    results_abs_reshaped = np.zeros((len(t_eval), N_ages, 5))
    for k in range(5):
        results_abs_reshaped[:, :, k] = results_abs_flat[:, k*N_ages : (k+1)*N_ages]

    # Cumulative cost is the last column (index 5*N_ages)
    cost_series = results_abs_flat[:, 5*N_ages]
    total_cost = cost_series[-1]

    return sol.t, results_abs_reshaped, total_cost


# 3) Evaluate objective for one combination (maximize lives saved per dollar, incl. R&D)
def evaluate_timing_combo(t_vax_start, t_anti_start, t_dist_start,
                          vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                          cost_vax, cost_anti, cost_dist,
                          baseline_final_deaths,
                          cost_RnD_vax, cost_RnD_anti):
    """
    Runs a single simulation and returns:
      - final_total_deaths,
      - peak_total_infections,
      - total_cost_flow: cumulative cost from flows C(T),
      - total_cost_with_RnD: flow cost + R&D,
      - lives_saved,
      - lives_per_dollar.
    """
    t, results, total_cost_flow = run_single_timing_sim(
        t_vax_start, t_anti_start, t_dist_start,
        vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
        cost_vax, cost_anti, cost_dist
    )

    # results: (Time x Age x Compartment)
    total_I = results[:, :, 2].sum(axis=1)  # Infectious across all ages
    total_D = results[:, :, 4].sum(axis=1)  # Deaths across all ages

    final_total_deaths = total_D[-1]
    peak_total_infections = total_I.max()

    lives_saved = baseline_final_deaths - final_total_deaths

    # --- Up-front R&D costs (only if the intervention is actually deployed) ---
    horizon_end = t_span[1]

    used_vax = (t_vax_start < horizon_end) and np.any(vax_rate > 0)
    used_anti = (t_anti_start < horizon_end) and np.any(antiviral_efficacy > 0)

    RnD_cost = 0.0
    if used_vax:
        RnD_cost += cost_RnD_vax
    if used_anti:
        RnD_cost += cost_RnD_anti

    total_cost_with_RnD = total_cost_flow + RnD_cost

    if total_cost_with_RnD > 0 and lives_saved > 0:
        lives_per_dollar = lives_saved / total_cost_with_RnD
    else:
        lives_per_dollar = -np.inf  # effectively discard non-cost-effective or no-benefit combos

    return (final_total_deaths,
            peak_total_infections,
            total_cost_flow,
            total_cost_with_RnD,
            lives_saved,
            lives_per_dollar)


# 4) Grid search over all three start times, maximizing lives saved per dollar
def joint_timing_grid_search(vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                             t_vax_values, t_anti_values, t_dist_values,
                             cost_vax, cost_anti, cost_dist,
                             baseline_final_deaths,
                             cost_RnD_vax, cost_RnD_anti):
    """
    Explores a 3D grid of start times:
      t_vax_values   : array of vaccination start times (days)
      t_anti_values  : array of antiviral start times (days)
      t_dist_values  : array of distancing start times (days)

    Returns:
      best_result: dict with best combo (max lives per dollar) and metrics
      deaths_grid: 3D array of final total deaths for each combo
    """
    deaths_grid = np.zeros((len(t_vax_values), len(t_anti_values), len(t_dist_values)))

    best = {
        "t_vax_start": None,
        "t_anti_start": None,
        "t_dist_start": None,
        "final_deaths": np.inf,
        "total_cost_flow": np.inf,
        "total_cost_with_RnD": np.inf,
        "lives_saved": -np.inf,
        "lives_per_dollar": -np.inf,
        "peak_infections": np.inf
    }

    for i, tv in enumerate(t_vax_values):
        for j, ta in enumerate(t_anti_values):
            for k, td in enumerate(t_dist_values):
                (final_deaths,
                 peak_I,
                 total_cost_flow,
                 total_cost_with_RnD,
                 lives_saved,
                 lives_per_dollar) = evaluate_timing_combo(
                    tv, ta, td,
                    vax_rate, vax_efficacy, antiviral_efficacy, distancing_factor,
                    cost_vax, cost_anti, cost_dist,
                    baseline_final_deaths,
                    cost_RnD_vax, cost_RnD_anti
                )

                deaths_grid[i, j, k] = final_deaths

                # Main objective: maximize lives saved per dollar
                if lives_per_dollar > best["lives_per_dollar"]:
                    best.update({
                        "t_vax_start": tv,
                        "t_anti_start": ta,
                        "t_dist_start": td,
                        "final_deaths": final_deaths,
                        "total_cost_flow": total_cost_flow,
                        "total_cost_with_RnD": total_cost_with_RnD,
                        "lives_saved": lives_saved,
                        "lives_per_dollar": lives_per_dollar,
                        "peak_infections": peak_I
                    })

    return best, deaths_grid


# 5) Example usage: define intervention intensities, costs, and timing grid, then run the search

# Use your "Scenario B" style intensities as the fixed strengths:
vax_rate_opt    = np.array([0.01, 0.005, 0.015])  # per-day vax rate by age
vax_eff_opt     = 0.8                             # vaccine efficacy (scalar)
anti_eff_opt    = np.array([0.0, 0.0, 0.5])       # antivirals mainly for elderly
dist_factor_opt = np.array([0.0, 0.2, 0.0])       # distancing mainly for adults

# Cost parameters (illustrative, in arbitrary currency units)
cost_vax = np.array([20.0, 20.0, 20.0])   # $ per vaccinated person (children, adults, elderly)
cost_anti = np.array([50.0, 75.0, 100.0]) # $ per treated infectious person
cost_dist = np.array([50.0, 150.0, 50.0]) # $ per person-day at full distancing

# R&D costs (illustrative order-of-magnitude)
cost_RnD_vax  = 5e9   # e.g., $5B to develop and deploy a new vaccine
cost_RnD_anti = 2e9   # e.g., $2B to develop antivirals

# Grids of start times (you can refine these later)
t_vax_values   = np.arange(0, 181, 30)   # 0, 30, 60, 90, 120, 150, 180
t_anti_values  = np.arange(0, 181, 30)
t_dist_values  = np.arange(0, 181, 30)

# Baseline scenario: no interventions, to compute baseline deaths
vax_zero = np.zeros_like(vax_rate_opt)
anti_zero = np.zeros_like(anti_eff_opt)
dist_zero = np.zeros_like(dist_factor_opt)

t_base, results_base, cost_base = run_single_timing_sim(
    t_vax_start=300.0,   # starts after horizon -> effectively off
    t_anti_start=300.0,
    t_dist_start=300.0,
    vax_rate=vax_zero,
    vax_efficacy=vax_eff_opt,  # efficacy irrelevant when vax_rate=0
    antiviral_efficacy=anti_zero,
    distancing_factor=dist_zero,
    cost_vax=cost_vax,
    cost_anti=cost_anti,
    cost_dist=cost_dist
)
baseline_final_deaths = results_base[-1, :, 4].sum()

best_combo, deaths_grid = joint_timing_grid_search(
    vax_rate_opt, vax_eff_opt, anti_eff_opt, dist_factor_opt,
    t_vax_values, t_anti_values, t_dist_values,
    cost_vax, cost_anti, cost_dist,
    baseline_final_deaths,
    cost_RnD_vax, cost_RnD_anti
)

print("\n=== Best Timing Combination (max lives saved per dollar) ===")
print(f"Vaccination start time: {best_combo['t_vax_start']} days")
print(f"Antiviral start time : {best_combo['t_anti_start']} days")
print(f"Distancing start time: {best_combo['t_dist_start']} days")
print(f"Final total deaths   : {best_combo['final_deaths']:.0f}")
print(f"Baseline deaths      : {baseline_final_deaths:.0f}")
print(f"Lives saved          : {best_combo['lives_saved']:.1f}")
print(f"Flow cost only       : {best_combo['total_cost_flow']:.2e}")
print(f"Total cost + R&D     : {best_combo['total_cost_with_RnD']:.2e}")
print(f"Lives per dollar     : {best_combo['lives_per_dollar']:.3e}")
print(f"Peak total infections: {best_combo['peak_infections']:.0f}")

# Re-run the best scenario to inspect full trajectories:
t_best, results_best, cost_best = run_single_timing_sim(
    best_combo['t_vax_start'],
    best_combo['t_anti_start'],
    best_combo['t_dist_start'],
    vax_rate_opt, vax_eff_opt, anti_eff_opt, dist_factor_opt,
    cost_vax, cost_anti, cost_dist
)


# 6) Visualization: Show epidemic curves with best start times

def plot_best_timing_summary(t, results, best_combo):
    """
    t: array of time points
    results: (Time x Age x Compartment)
    best_combo: dict from joint_timing_grid_search with optimal start times
    """
    total_I = results[:, :, 2].sum(axis=1)  # total infectious over time
    total_D = results[:, :, 4].sum(axis=1)  # total deaths over time

    fig, ax1 = plt.subplots(1, 1, figsize=(10, 6))

    # Epidemic curves
    ax1.plot(t, total_I, label='Total Infectious (I)', linewidth=2)
    ax1.plot(t, total_D, label='Total Deaths (D)', linewidth=2)

    # Draw vertical lines for best start times
    ax1.axvline(best_combo['t_vax_start'],  linestyle='--', alpha=0.8,
                label=f"Vax start = {best_combo['t_vax_start']} d")
    ax1.axvline(best_combo['t_anti_start'], linestyle='--', alpha=0.8,
                label=f"Antivirals start = {best_combo['t_anti_start']} d")
    ax1.axvline(best_combo['t_dist_start'], linestyle='--', alpha=0.8,
                label=f"Distancing start = {best_combo['t_dist_start']} d")

    ax1.set_title('Best-Timing Scenario: Epidemic Curves & Intervention Onset', fontsize=12)
    ax1.set_xlabel('Time (days)')
    ax1.set_ylabel('Number of individuals')
    ax1.legend(loc='upper left', fontsize=8)
    ax1.grid(True, linestyle=':', alpha=0.6)

    plt.tight_layout()
    plt.savefig("best_lives_per_dollar")
    plt.show()

# Call the plotting function for the best combo
plot_best_timing_summary(t_best, results_best, best_combo)


# 7) Sensitivity analysis: vary vaccine R&D cost and visualize impact

# Fix antiviral R&D cost for this experiment (reuse cost_RnD_anti)
cost_RnD_anti_fixed = cost_RnD_anti

# Sweep vaccine R&D cost over a range
vax_RnD_grid = np.linspace(0, 10e9, 6)  # 0, 2B, 4B, ..., 10B

best_lives_saved = []
best_lives_per_dollar = []
best_vax_start_times = []

for cR in vax_RnD_grid:
    best_tmp, _ = joint_timing_grid_search(
        vax_rate_opt, vax_eff_opt, anti_eff_opt, dist_factor_opt,
        t_vax_values, t_anti_values, t_dist_values,
        cost_vax, cost_anti, cost_dist,
        baseline_final_deaths,
        cost_RnD_vax=cR,
        cost_RnD_anti=cost_RnD_anti_fixed
    )
    best_lives_saved.append(best_tmp["lives_saved"])
    best_lives_per_dollar.append(best_tmp["lives_per_dollar"])
    best_vax_start_times.append(best_tmp["t_vax_start"])

best_lives_saved = np.array(best_lives_saved)
best_lives_per_dollar = np.array(best_lives_per_dollar)
best_vax_start_times = np.array(best_vax_start_times)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: lives saved vs R&D cost
ax = axes[0]
ax.plot(vax_RnD_grid, best_lives_saved, marker='o')
ax.set_xlabel("Vaccine R&D cost")
ax.set_ylabel("Best lives saved (vs baseline)")
ax.set_title("Lives Saved vs Vaccine R&D Cost")
ax.grid(True, linestyle=':', alpha=0.6)

# Plot 2: lives per dollar vs R&D cost
ax = axes[1]
ax.plot(vax_RnD_grid, best_lives_per_dollar, marker='s')
ax.set_xlabel("Vaccine R&D cost")
ax.set_ylabel("Best lives per dollar")
ax.set_title("Cost-Effectiveness vs Vaccine R&D Cost")
ax.grid(True, linestyle=':', alpha=0.6)

# Plot 3: optimal vaccine start time vs R&D cost
ax = axes[2]
ax.plot(vax_RnD_grid, best_vax_start_times, marker='^')
ax.set_xlabel("Vaccine R&D cost")
ax.set_ylabel("Optimal vaccine start time (days)")
ax.set_title("Optimal Timing vs Vaccine R&D Cost")
ax.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```
